```python
from selenium import webdriver
import requests
import time
import os, json
from os import path
```

```python
driver = webdriver.Chrome('/Users/bhargav/Developer/PycharmProjects/python-internship/chromedriver')
driver.get('https://gomechanic.in/pune')
```

```python
links = {}
for i in range(1, 42):
    link = driver.find_element_by_xpath(f'//*[@id="app"]/div[2]/div[1]/div[1]/div/div/div/div/div[3]/div/a[{i}]').get_attribute('href')
    brand = driver.find_element_by_xpath(f'//*[@id="app"]/div[2]/div[1]/div[1]/div/div/div/div/div[3]/div/a[{i}]/div/div').text
    links[brand] = link
```

```python
models = {}
for link in links:
    print(links[link], link, sep=' ')
    driver.get(links[link])
    time.sleep(2)
    models_len = len(driver.find_elements_by_xpath('//*[@id="app"]/div[1]/div/div[3]/div/div/div/div/div[3]/div/a'))
    print(models_len)
    dic = {}
    for i in range(1, models_len + 1):
        image = driver.find_element_by_xpath(f'//*[@id="app"]/div[1]/div/div[3]/div/div/div/div/div[3]/div/a[{i}]/div/img')
        model = driver.find_element_by_xpath(f'//*[@id="app"]/div[1]/div/div[3]/div/div/div/div/div[3]/div/a[{i}]/div/div').text
        driver.execute_script("arguments[0].scrollIntoView();", image)
        print(image.get_attribute('src'), model)
        dic[model] = image.get_attribute('src')
    models[link] = dic
```

```python
for brand in models:
    os.chdir(f'/Users/bhargav/Developer/PycharmProjects/python-internship/imgs/{brand}')
    for model in models[brand]:
        print(path.join(os.getcwd(), model) + '.jpeg')
        image_path = path.join(os.getcwd(), model) + '.jpeg'
        image_data = requests.get(models[brand][model]).content
        print(models[brand][model], brand, model)
        file = open(image_path, 'wb')
        file.write(image_data)
    os.chdir('/')
```
